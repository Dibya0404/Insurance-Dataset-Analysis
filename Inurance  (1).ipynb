```python
plt.scatter(med['age'], med['charges'])
plt.title("Age vs Insurance Charges")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

```python
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=med)
plt.title("BMI vs Charges (Smoker & Non-Smoker)")
plt.show()
```

```python
sns.boxplot(x='smoker', y='charges', data=med)
plt.title("Smoker vs Charges")
plt.show()
```

```python
sns.countplot(x='region', data=med)
plt.title("Number of People per Region")
plt.show()
```

```python
plt.hist(med['bmi'], bins=20)
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(med['charges'], bins=30, kde=True)
plt.title("Distribution of Insurance Charges")
plt.show()
```

```python
sns.barplot(x='children', y='charges', data=med)
plt.title("Children vs Charges")
plt.show()
```

```python
sns.barplot(x='region', y='charges', data=med, estimator=np.mean)
plt.title("Average Charges by Region")
plt.show()
```

```python
med['age_group'] = pd.cut(med['age'], bins=[17,25,35,45,55,65], labels=['18-25','26-35','36-45','46-55','56-65'])
sns.barplot(x='age_group', y='charges', data=med)
plt.title("Average Charges by Age Group")
plt.show()
```

```python
sns.violinplot(x='sex', y='charges', data=med)
plt.title("Gender vs Insurance Charges")
plt.show()
```

```python
sns.pairplot(med, hue='smoker')
plt.suptitle("Pairplot of All Features", y=1.02)
plt.show()
```
